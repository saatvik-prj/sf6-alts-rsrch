```python
import pandas as pd
from scipy.optimize import minimize

# Load data
data = pd.read_csv("updated_gas_properties.csv")

# Objective function: Minimize GWP while maximizing Dielectric Strength
def objective(x):
    gwp = sum(x * data['GWP'])
    strength = sum(x * data['Dielectric Strength (kV/cm)'])
    return gwp - strength * 0.1  # Adjust weight as needed

# Constraints: Sum of proportions should be 1
constraints = [{'type': 'eq', 'fun': lambda x: sum(x) - 1}]

# Bounds: Each proportion between 0 and 1
bounds = [(0, 1) for _ in range(len(data))]

# Initial guess
x0 = [1 / len(data)] * len(data)

# Solve
result = minimize(objective, x0, bounds=bounds, constraints=constraints)
print("Optimal mix:", result.x)
```
